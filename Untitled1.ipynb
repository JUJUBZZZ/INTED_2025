```python
#1 – Em um país fictício, o direito ao voto começa aos 16 anos. Crie um programa que receba a idade de uma pessoa e determine, usando apenas uma condição simples if, se ela já pode votar.
idade = int(input("Digite sua idade: "))

pode_votar = "pode votar" if idade >= 16 else "não pode votar"

print(f"Você {pode_votar}.")
```

```python
#2 – Uma escola deseja automatizar a verificação do desempenho dos alunos. Leia duas notas, calcule a média e utilize uma estrutura if…else para informar se o aluno está aprovado (média ≥ 7) ou reprovado.
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))

media = (nota1 + nota2) / 2

if media >= 7:
    print(f"Média = {media:.1f}. Aluno APROVADO.")
else:
    print(f"Média = {media:.1f}. Aluno REPROVADO.")
```

```python
#3 – Um sistema meteorológico quer classificar a sensação térmica do dia. A partir de uma temperatura informada, determine se está fria, agradável, quente ou muito quente usando if…elif…else.
temperatura = float(input("Informe a temperatura do dia (°C): "))

if temperatura < 15:
    print("A sensação térmica está *fria*.")
elif temperatura < 25:
    print("A sensação térmica está *agradável*.")
elif temperatura < 35:
    print("A sensação térmica está *quente*.")
else:
    print("A sensação térmica está *muito quente*.")
```
